## 해상도 컬럼 뽑아서 테이블 생성

```python
import pandas as pd
import re

# 원본 TV csv 경로
input_path = "C:/project_skn/project4/4th_project/products/data/preprocessing/TV/lge_tv_all_products.csv"

# 저장할 csv 경로
output_path = "ScreenResolution.csv"

# 원본 읽기
tv_df = pd.read_csv(input_path)

# 해상도 컬럼만 Series로 추출
resolution_series = tv_df["해상도"].dropna()

rows = []
seen = set()

for value in resolution_series:
    value = str(value).strip()

    # 예: HD (1,366 x 768)
    match = re.match(r"(.+?)\s*\(([\d,]+)\s*[xX×]\s*([\d,]+)\)", value)

    if not match:
        continue

    name = match.group(1).strip()
    width = int(match.group(2).replace(",", ""))
    height = int(match.group(3).replace(",", ""))

    key = (name, width, height)

    # 중복 제거
    if key not in seen:
        seen.add(key)
        rows.append({
            "name": name,
            "width": width,
            "height": height
        })

# DataFrame 생성
resolution_df = pd.DataFrame(rows)

# resol_code 생성: RES000, RES001, ...
resolution_df.insert(
    0,
    "resol_code",
    [f"RES{i:03d}" for i in range(len(resolution_df))]
)

# 타입 지정
resolution_df = resolution_df.astype({
    "resol_code": "string",
    "name": "string",
    "width": "int",
    "height": "int"
})

# csv 저장
resolution_df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(resolution_df)
print(f"저장 완료: {output_path}")
```
